```python
def filter_data(data, valid_country_codes, variables):
    # Create a boolean mask for each condition
    mask_country = data['REF_AREA'].isin(valid_country_codes)
    mask_indicator = data['INDICATOR'].isin(variables)
    mask_year = data['original_period'].str[:4].astype(int) >= 1970

    # Combine all conditions and filter the data
    filtered_data = data.loc[mask_country & mask_indicator & mask_year]

    return filtered_data
```

```python
def restructure_data(df):
    # extracting the year
    df['year'] = df['original_period'].apply(lambda x: x[:4]).astype('int')
    #extracting the quarter
    df['quarter'] = df['original_period'].apply(lambda x: x.split('-Q')[1])

    # putting  together year and quarter for the original period
    df['year_quarter'] = df['year'].astype(str) + 'Q' + df['quarter']

    # reformatting the series code
    df.loc[:, 'series_code'] = df['series_code'].apply(lambda x: '.'.join(x.split('.')[1:]))
    df.loc[:, 'description'] = df['Indicator'] + ', ' + df['Reference Area'] + ', ' + df['Frequency']

    # calculating the max period for the year and for quarter for each series_code
    max_period = pd.DataFrame(df.groupby('series_code')['year_quarter'].max())
    max_period.reset_index(inplace=True)

    # merging the max period to the data frame
    df = pd.merge(df, max_period, how='left', on='series_code', suffixes=('', '_max'))

    # renaming column names and selecting the desired columns
    df = df.loc[:, ['series_code', 'description', 'year_quarter', 'year_quarter_max', 'value']]
    df.columns = ['code', 'description', 'period', 'cutoff', 'value']

    return df
```

```python
def pivot_table(df):

    df['period'] = df['period'].astype(str)

    pivot_df = df.set_index(['code', 'description', 'cutoff', 'period'])['value'].unstack('period')

    # reset index
    pivot_df.reset_index(inplace=True)

    return pivot_df
```
